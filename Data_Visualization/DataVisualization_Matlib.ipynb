```python
df=pd.DataFrame({"StudyHours":[10,11.5,9,6,5,8,7,12,10,11,25],
                "Grade":[80,90,75,55,50,60,58,96,85,89,225],
                "Gender":["M","M","F","M","F","F","M","F","F","M","M"],
                "n_ass":[3,4,2,1,1,2,3,5,4,5,15]})
df
```

```python
import matplotlib.pyplot as plt
```

### Histogram

```python
plt.hist(df["Grade"],bins=10,range=(0,100))
plt.xticks(list(range(0,101,20)))
plt.xlabel("Grade")
plt.ylabel("No. of students")
plt.title("Report_Card")

plt.show()
```

### Box Plot

```python
plt.boxplot(df["Grade"])
plt.show()
```

```python
q3=df["Grade"].quantile(0.75)
q1=df["Grade"].quantile(0.25)
iqr=q3-q1
upper_limit=q3+(1.5*iqr)
lower_limit=q1-(1.5*iqr)
df[(df["Grade"]<lower_limit) | (df["Grade"]>upper_limit)]
# print(q1,q3,upper_limit,lower_limit)
```

### Scatter Plot

```python
plt.scatter(x=df['StudyHours'],y=df['Grade'])
plt.xlabel('Study Hours')
plt.ylabel('Grade')
plt.show()
```

### Pie Chart

```python
plt.pie(x=df['Gender'].value_counts(),labels=df['Gender'].value_counts().index.to_list(),autopct='%0.1f%%')  # autopct means upto how much decimal places you wish to have the percentage
plt.show()
```

### Bar Plot

```python
x=df['Gender'].value_counts().index.tolist()
y=df['Gender'].value_counts()
plt.bar(x,y)
plt.xlabel("Categories of Gender Variable")
plt.ylabel("Count of each category ")
plt.show()
```

```python
plt.barh(x,y)
plt.xlabel("no.of students")
plt.show()
```

### Stacked Barplot

```python
company=['gogl','msft','amzn','fb']
turnover=[101,19,125,90]
profit=[80,6,100,52]
plt.bar(company,turnover,label="turnover")
plt.bar(company,profit,label="profit")
plt.xlabel("company")
# plt.yticks(np.arange(0,200,50))
plt.legend()
plt.show()
```

### Unstacked Barplot

```python
company=['gogl','msft','amzn','fb']
turnover=[101,19,125,190]
profit=[80,6,100,52]
plt.bar([0,1,2,3],turnover,width=0.4,label="turnover")
plt.bar([0.4,1.4,2.4,3.4],profit,width=0.4,label="profit")

# plt.xlabel('company')

# plt.xticks([0,1,2,3],company)


plt.legend()
plt.show()

```

### Subplots

```python
company=['gogl','msft','amzn','fb']
q1=[45,85,74,36]
q2=[56,41,35,85]
q3=[85,47,25,69]
q4=[58,74,14,58]

plt.figure(figsize=(10,12))

plt.subplot(2,2,1)
plt.bar(company,q1)

plt.subplot(2,2,2)
plt.bar(company,q2)

plt.subplot(2,2,3)
plt.bar(company,q3)

plt.subplot(2,2,4)
plt.bar(company,q4)

plt.show()
```
